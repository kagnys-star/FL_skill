```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import umap
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
```

```python
with np.load('/home/kangys/workspace/FL_skill/spirl/experiments/normal.npz') as data:
    latent_arr = data['arr_0']
    label_arr = data['arr_1']
```

```python
pca = PCA(n_components=2)
data_pca = pca.fit_transform(latent_arr)
```

```python
plt.figure(figsize=(10, 8))
for i in np.unique(label_arr):
    indices = np.where(label_arr == i)
    plt.scatter(data_pca[indices, 0], data_pca[indices, 1], label=f'Label {i}')
plt.title('2D PCA Projection of the Data')
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.legend()
plt.show()
```

```python
tsne = TSNE(n_components=2, random_state=42)
data_tsne = tsne.fit_transform(latent_arr)
```

```python
plt.figure(figsize=(10, 8))
for i in np.unique(label_arr):
    indices = np.where(label_arr == i)
    plt.scatter(data_tsne[indices, 0], data_tsne[indices, 1], label=f'Label {i}')
plt.title('2D TSNE Projection of the Data')
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.legend()
plt.show()
```

```python
umap_model = umap.UMAP(n_components=2, random_state=42)
data_umap = umap_model.fit_transform(latent_arr)
```

```python
plt.figure(figsize=(10, 8))
for i in np.unique(label_arr):
    indices = np.where(label_arr == i)
    plt.scatter(latent_arr[indices, 0], latent_arr[indices, 1], label=f'Label {i}')
plt.title('2D UMAP Projection of the Data')
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.legend()
plt.show()
```

```python
lda = LDA(n_components=2)
data_lda = lda.fit_transform(latent_arr, label_arr)
```

```python
plt.figure(figsize=(10, 8))
for i in np.unique(label_arr):
    indices = np.where(label_arr == i)
    plt.scatter(data_lda[indices, 0], data_lda[indices, 1], label=f'Label {i}')
plt.title('2D UMAP Projection of the Data')
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.legend()
plt.show()
```
